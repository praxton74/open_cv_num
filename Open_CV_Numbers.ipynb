```python
from keras.layers import Dense, Conv2D,AveragePooling2D,Flatten
```

```python
from keras.models import Sequential
model=Sequential()
```

```python
model.add(Conv2D(filters = 4 , kernel_size = (5,5), input_shape=(28,28,1),activation='relu'))
```

```python
model.add(AveragePooling2D(pool_size=(2,2)))
```

```python
model.add(Conv2D(filters = 4 , kernel_size = (5,5),activation='relu'))
```

```python
model.add(AveragePooling2D(pool_size=(2,2)))
```

```python
model.add(Flatten())
```

```python
model.add(Dense(units = 10, activation='softmax'))
```

```python
model.compile(loss="sparse_categorical_crossentropy", optimizer='adam',metrics=['acc'])
```

```python
model.summary()
```

```python
model.fit(X_train,y_train,epochs=1,batch_size=1)
```

```python
plt.imshow(X_test[88])
```

```python
test=X_test[88].reshape(-1,28,28,1)
```

```python
y_predict = np.argmax(model.predict(test), axis=-1)
```
